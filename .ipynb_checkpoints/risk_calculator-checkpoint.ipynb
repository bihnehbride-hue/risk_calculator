```python
_citylist={"new york":10,"texas":16,"los angeles":25,"chicago":32,"W.dc":18}
userlocation = []
for n in range(5):
    location=input("Enter your location: ")
    userlocation.append(location)
for i in userlocation:
    score=_citylist[i]
    if score > 14:
        print(i, ":","safe")
    else:
        print(i, ":","risk")
```
